```python
comb = pd.concat([train,test], ignore_index=True, sort=False)
comb.shape
```

```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r :
        input_txt = re.sub(re.escape(i), " ", input_txt)
    return input_txt    
```

```python
#removing twitter handles
comb['tidy_tweets'] = np.vectorize(remove_pattern)(comb['tweet'], "@[\w]*")
comb.head()
```

```python
# removing punctuations
# comb['tidy_tweets'] = comb['tidy_tweets'].str.replace("[^a-zA-Z#]", " " )
comb['tidy_tweets'] = np.vectorize(remove_pattern)(comb['tidy_tweets'], "[^a-zA-Z#]")

comb.head(10)
```

```python
# removing short words
comb['tidy_tweets']= comb['tidy_tweets'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
comb.head(10)
```

```python
# text normalization
tokenized_tweet = comb['tidy_tweets'].apply(lambda x: x.split())
#tokenized_tweet.head()
from nltk.stem.porter import *
stem = PorterStemmer()
tokenized_tweet = tokenized_tweet.apply(lambda x : [stem.stem(i) for i in x]) 

#joining tokens again
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ' '.join(tokenized_tweet[i])
comb['tidy_tweets'] = tokenized_tweet
comb.head(10)
```

```python
# Understanding impacts of hashtags on tweets setiment
def hashtag_extract(x):
    hashtags=[]
    for i in x:
        ht = re.findall(r"#(\w+)", i)
        hashtags.append(ht)
    return hashtags

# extracting hashtgs from positive tweets
pos_tags = hashtag_extract(comb['tidy_tweets'][comb['label']==0])
#extracting hashtags from negative tweets
neg_tags = hashtag_extract(comb['tidy_tweets'][comb['label']==1])
#unnesting list
pos_tags = sum(pos_tags,[])
neg_tags = sum(neg_tags,[])

    
```

```python
# Positive tweets
a = nltk.FreqDist(pos_tags)
d = pd.DataFrame({'Hashtag':list(a.keys()), 'Count':list(a.values())})
#selecting top 20 most frequencies hastags
d = d.nlargest(columns='Count', n=20)
plt.figure(figsize=(16,5))
plt.bar(d['Hashtag'],d['Count'])
plt.ylabel('Count')
plt.xlabel('Hashtag')
plt.title('Top 20 Hashtags in Positive Tweets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Negative tweets
b = nltk.FreqDist(neg_tags)
d = pd.DataFrame({'Hashtag':list(b.keys()), 'Count':list(b.values())})
#selecting top 20 most frequencies hastags
d = d.nlargest(columns='Count', n=20)
plt.figure(figsize=(16,5))
plt.bar(d['Hashtag'],d['Count'])
plt.ylabel('Count')
plt.xlabel('Hashtag')
plt.title('Top 20 Hashtags in Negative Tweets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# constructing text features
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer 
import gensim
vectorizer = CountVectorizer(max_df=0.9, min_df=2, max_features=1000, stop_words='english')
bow = vectorizer.fit_transform(comb['tidy_tweets'])
bow.shape
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.9, min_df=2, max_features=1000, stop_words='english')
tfidf = tfidf_vectorizer.fit_transform(comb['tidy_tweets'])
tfidf.shape
```

```python
# training word2vec model
tokenized_tweet = comb['tidy_tweets'].apply(lambda x: x.split())

w2v = gensim.models.Word2Vec(
    tokenized_tweet,
    vector_size=100 ,#no of features,
    window=3,
    min_count=2,
    sg=1,
    hs=0,
    negative=10,
    workers=1,
    seed=34)
w2v.train(tokenized_tweet, total_examples = len(comb['tidy_tweets']),epochs=20)
```
